```python
import pandas as pd
import numpy as np
import math
```

```python
filename = '../data/data.txt'  
with open(filename) as fn:  
    # Read each line
    ln = fn.readline()
    # Keep count of lines
    lncnt = 0
    validCounter=0
    totalCounter=0
    counter=0
    byr_flag = 0
    iyr_flag = 0
    eyr_flag = 0
    hgt_flag = 0
    hcl_flag = 0
    ecl_flag = 0
    pid_flag = 0
    cid_flag = 0
    while ln:
        x = ln.split(" ")
#         print(x[0]=="\n")
        print("Line {}: {}".format(lncnt, ln.strip()))
        if (x[0]=="\n")==False:
            counter=counter+1
#             print(counter)
            if counter!=0:
                for i in range(0,len(x)):
                    b = x[i].split(":")
#                     print(b[0])
                    if b[0] == "byr":
                        byr_flag=byr_flag+1
                    elif b[0] == "iyr":
                        iyr_flag=iyr_flag+1
                    elif b[0] == "eyr":
                        eyr_flag=eyr_flag+1
                    elif b[0] == "hgt":
                        hgt_flag=hgt_flag+1
                    elif b[0] == "hcl":
                        hcl_flag=hcl_flag+1
                    elif b[0] == "ecl":
                        ecl_flag=ecl_flag+1
                    elif b[0] == "pid":
                        pid_flag=pid_flag+1
                    elif b[0] == "cid":
                        cid_flag=cid_flag+1
        else:
            test = byr_flag*iyr_flag*eyr_flag*hgt_flag*hcl_flag*ecl_flag*pid_flag
            print("test Value:",test)
            if test ==1:
                print("valid")
                validCounter= validCounter+1
            totalCounter = totalCounter +1
            byr_flag = 0
            iyr_flag = 0
            eyr_flag = 0
            hgt_flag = 0
            hcl_flag = 0
            ecl_flag = 0
            pid_flag = 0
            cid_flag = 0
            print("valid counter:",validCounter)
            print("total counter :", totalCounter)

        ln = fn.readline()
        lncnt += 1
    print("valid counter:",validCounter)
    print("total counter :", totalCounter)
```
