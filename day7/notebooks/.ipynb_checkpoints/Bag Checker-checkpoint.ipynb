```python
import numpy as np
import pandas as pd
import re
import math
```

```python
filename = "../data/data.txt"
colnames = ["original","content2","content1","content1_num","content1_color","content2_num","content2_color","original_num","original_color"]
df = pd.read_csv(filename, sep=",", header=None,names=colnames)
# print(df)
df[["original","content1"]] = df['original'].str.split("contain",expand=True)
# print(df)
rows = df.shape[0]

def split_string_col(df,col_name,row):
    a = df[col_name][row]
    a2 = isinstance(a, float)
    if a2==False:
        b = a
        c = b.split(" ")
        if not c[0]=="no" :
            df[col_name+"_color"][row] = c[2]+c[3]
            df[col_name+"_num"][row] = c[1]
        else:
            df[col_name+"_color"][row] = "no"
            df[col_name+"_num"][row] = "no"
    else:
        df[col_name][row] = "no"
        df[col_name+"_color"][row] = "no"
        df[col_name+"_num"][row] = "no"
    return(df)

def split_string_col2(df,col_name,row):
    a = df[col_name][row]
    a2 = isinstance(a, float)
    print(a2)
    if a2==False:
        b = a
        c = b.split(" ")
        if not c[0]=="no" :
            df[col_name+"_color"][row] = c[0]+c[1]
            df[col_name+"_num"][row] = c[1]
        else:
            df[col_name+"_color"][row] = "no"
            df[col_name+"_num"][row] = "no"
    else:
        df[col_name][row] = "no"
        df[col_name+"_color"][row] = "no"
        df[col_name+"_num"][row] = "no"
    return(df)

for row in range(0,rows):
    col_name="content1"
    df_c1 = split_string_col(df,col_name,row)


for row in range(0,rows):
    col_name="content2"
    df_c2 = split_string_col(df_c1,col_name,row)


for row in range(0,rows):
    col_name="original"
    df_final = split_string_col2(df_c2,col_name,row)
```
